In this exercise you are going to analyze a yelp review, and try to determine the number of stars were given.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

yelp = pd.read_csv('yelp.csv')
```

Using CountVectorizor make the X the text, and make y the stars.  You will need to do a fit_transform on X.

```python
# i should take the text review as a feature and stars as a target
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

X = yelp['text']
y = yelp['stars']

X = cv.fit_transform(X)
```

Do a train test split and a Naive Bayes on your data.

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

#i split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=87)

# then creating my model
nb = MultinomialNB()

# train model
nb.fit(X_train, y_train)
```

Determine the accuracy of your model by printing the classification report

```python
from sklearn.metrics import classification_report

predictions = nb.predict(X_test)

print(classification_report(y_test, predictions))
```

```python
yelp1or5 = yelp[(yelp['stars'] == 5) | (yelp['stars'] == 1)]
```

Let's try to filter our data to the more extreme responses.  Make a new dataframe using only responses that are 1 star or 5 stars. <br><br>
yelp1or5 = yelp[(yelp['stars']==5) | (yelp['stars']==1)]

Redo the above steps using yelp1or5.  What is the accuracy of your model now? Print a classification report.  Why do you think there is a change?

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

X = yelp1or5['text']
y = yelp1or5['stars']

cv = CountVectorizer()

X = cv.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=87)

nb = MultinomialNB()

nb.fit(X_train, y_train)

predictions = nb.predict(X_test)

print(classification_report(y_test, predictions))
```

```python
# The model is not very well when predicting all 5 star categories
# because categorizing reviews into different 5 star is more difficult!
# when i sert dataset to only 1 star and 5 star reviews the model worked  better
# these two classes are more extreme and easier to separate based on the review text
```
